```python
#we use feature group method of folium to add extra features to map

m=folium.Map(location=[18.36,73.47],zoom_start=7,tiles="Stamen Terrain")
fg=folium.FeatureGroup(name="My Map")
fg.add_child(folium.Marker(location=[18.36,73.47],popup="I_live_here",icon=folium.Icon(color='green')))
m.add_child(fg)
m.save("Map1.html")
```

```python
import folium
import pandas

data=pandas.read_csv("Volcanoes.txt")

lat=list(data["LAT"])
lon=list(data["LON"])
name=list(data["NAME"])
elev=list(data["ELEV"])

m=folium.Map(location=[38.58,-99.09],zoom_start=6,tiles="Stamen Terrain")
fg=folium.FeatureGroup(name="My Map")

for a,b,c,d in zip(lat,lon,name,elev) :
    if d<1500:
        fg.add_child(folium.Marker(location=[a,b],popup=c+" "+str( d),icon=folium.Icon(color='green')))
    
    elif d>3000:
        fg.add_child(folium.Marker(location=[a,b],popup=c+" "+str( d),icon=folium.Icon(color='red')))
        
    
    else:
        fg.add_child(folium.Marker(location=[a,b],popup=c+" "+str( d),icon=folium.Icon(color='orange')))
    

m.add_child(fg)
m.save("Map3.html")
```
