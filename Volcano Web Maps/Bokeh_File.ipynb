```python
#plotting the x and y values

from bokeh.plotting import figure
from bokeh.io import output_file,show

x=[1,2,3,4,5]
y=[6,7,8,9,10]

#note that these lists have same length otherwise bokeh will throw an error
output_file("line.html")#prepare the output file

f=figure()#create a figrue object
f.line(x,y)#create a line plot
show(f)#write the line plot in the figure object
```

```python
#you can add a title to  the plot,set the figure width ,height ,change font etc.

import pandas
from bokeh.plotting import figure
from bokeh.io import output_file,show
p=figure()
#p=figure(plot_width=500,plot_height=400 , tools='pan', logo=None)
p.title.text="Cool Data"
p.title.text_color="Gray"
p.title.text_font="times"
p.title.text_font_style="bold"
p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color=None
p.xaxis.axis_label="Date"
p.yaxis.axis_label="Intensity"

p.line([1,2,3],[4,5,6])
output_file("Graph.html")
show(p)
```

```python
#weather graph

import pandas

from bokeh.plotting import figure,output_file,show

df=pandas.read_excel("verlegenhuken.xls",sheet_name=0)

p=figure(plot_width=500,plot_height=400,tools='pan')

p.title.text="Temperature and air pressure"
p.title.text_color="Gray"
p.title.text_font="arial"
p.title.text_font_style="bold"
p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color=None
p.xaxis.axis_label="Temperatur(C)"
p.yaxis.axis_label="Pressure(hPa)"

p.circle(df["Temperature"],df["Pressure"],size=0.5)
output_file("Weather.html")
show(p)
```

```python
from bokeh.plotting import figure,output_file,show

p=figure(plot_width=500,plot_height=400,tools='pan,reset')

p.title.text="Earthquake"
p.title.text_color="Orange"
p.title.text_font="times"
p.title.text_font_style="italic"
p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color="Yellow"
p.xaxis.axis_label="Times"
p.yaxis.axis_label="Value"

p.circle([1,2,3,4,5],[5,6,5,5,3],size=[i*2 for i in [8,12,14,15,20]],color="red",alpha=0.5)

output_file("Scattering_plot.html")
show(p)
```
